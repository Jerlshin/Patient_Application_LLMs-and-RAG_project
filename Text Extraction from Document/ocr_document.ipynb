```python
import os
import PyPDF2  # Import PyPDF2 for PDF processing

def extract_text_from_pdf(pdf_path):
    # empty string to store extracted text
    text = ""
    pdf_reader = None

    try:
        # open the file 
        pdf_reader = PyPDF2.PdfReader(pdf_path)

    except PyPDF2.errors.PdfReaderError as e:
        print(f"Error reading PDF: {e}")
        return text

    # go through the each page in file 
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]

        # try to extract text directly 
        try:
            extracted_text = page.extract_text() # extract 
            text += extracted_text # add it to the total text
        except:
            pass
        
        # if text extraction fails, we should use OCR as that would be the image in the pdf 
        if not extracted_text:
            try:
                # temp image file path
                image_path = f"{pdf_path[:-4]}-{page_num}.jpg"
                # Open the PDF page as an image
                image = Image.open(image_path)

                # OCR 
                extracted_text = pytesseract.image_to_string(image)
                # append 
                text += extracted_text

            except Exception as e: 
                print(f"Error during OCR: {e}")
            finally:
                # clean 
                if os.path.exists(image_path):
                    os.remove(image_path)

    return text

pdf_path = "../Dummy_Medical_report/WM17S.pdf"
extracted_text = extract_text_from_pdf(pdf_path)

print(extracted_text)
```
